```python
with open('data.txt', 'rt', encoding='utf-8') as f:
    departments = {}
    for line in f:
        department_name = line.strip()
        employees_count = int(f.readline())
        employees = []
        for i in range(employees_count):
            emp = f.readline().strip()
            name, last_name, date_of_birth, city = emp.split(' | ')
            employees.append({
                'name': name,
                'last_name': last_name,
                'date': date_of_birth,
                'city': city
            })
        f.readline()
        departments[department_name] = employees
print(departments)
```
